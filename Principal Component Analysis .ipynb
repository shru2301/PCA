```python
df = pd.DataFrame(dataset.data,columns = dataset.feature_names)
```

```python
df
```

```python
df.describe()
```

```python
X = df
Y = dataset.target
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_s = scaler.fit_transform(X)
Xs = pd.DataFrame(X_s)
Xs.describe()
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xt, xt2, yt, yt2 = tts(X_s,Y,test_size=0.2,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(xt,yt)
```

```python
model.score(xt2,yt2)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(0.95)
```

```python
X_pca = pca.fit_transform(X)
```

```python
X_pca.shape
```

```python
X_pca1, X_pca2, Y_t, Y_tt = tts(X_pca,Y,test_size=0.2,random_state=30)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_pca1,Y_t)
model.score(X_pca2,Y_tt)
```

```python
pca = PCA(n_components = 15)
```

```python
X_pca = pca.fit_transform(X)
```

```python
X_pca1, X_pca2, Y_t, Y_tt = tts(X_pca,Y,test_size=0.2,random_state=30)
```

```python
model2 = LogisticRegression(max_iter=1000)
model2.fit(X_pca1,Y_t)
model2.score(X_pca2,Y_tt)
```
